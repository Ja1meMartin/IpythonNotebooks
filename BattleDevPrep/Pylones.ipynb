Given a list of Pylon heights. Look both left and right. You can only see a pylon if it is taller than all pylons between you and it. Return how many pylons can be seen from each pylon.

```python
folder = 'Pylones'

def read_input(folder, file_number=1):
    filename = folder + "/input" + str(file_number) + '.txt'
    with (open (filename) ) as name:
        lines = name.readlines()
        
    return [int(x) for x in lines[1:]]
    
def read_output(folder, file_number=1):
    filename = folder + "/output" + str(file_number) + '.txt'
    with (open (filename) ) as name:
        return [int(token)  for token in name.readline() if token != ' ']
    
print read_input(folder)
print read_output(folder)
```

```python
pylons = read_input(folder, 2)

print pylons

for n in range (len (pylons)):
    left, right = pylons[:n], pylons[n+1:]
    print left, right
```

```python
def look(values):
    
    pylons_seen = 0
    max_height = 0
    
    for height in values:
        if height > max_height:
            max_height = height
            pylons_seen += 1
            
    return pylons_seen

print "DESIRED RESULT", read_output(folder,2)

for n in range (len (pylons)):
    left, right = pylons[:n][::-1], pylons[n+1:]
    print left, look(left) 
    print right, look(right)
    print
```

```python
def check_pylons(pylons):
    views = []
    for n in range (len (pylons)):
        left, right = pylons[:n][::-1], pylons[n+1:]
        views.append(look(left) + look(right))
    return views

def test(folder):   
    for n in [1,2,3]:
        print check_pylons(read_input(folder, n)), read_output(folder, n)
        print all(zip(check_pylons(read_input(folder, n)), read_output(folder, n)))
        
test(folder)
```
