```python
def set_quote_strategy(text):
    import collections

    QuoteStrategy = collections.namedtuple('QuoteStrategy', 'lq rq')

    single = QuoteStrategy(lq="'", rq="'")
    double = QuoteStrategy(lq='``', rq="''")
    double_ord = QuoteStrategy(lq="''", rq="''")
    double_fancy = QuoteStrategy(lq='“', rq='”')
    fancy = QuoteStrategy(lq="‘", rq ='’')

    if "``" in text:
        return double
    
    if '”' in text:
        return double_fancy
    
    if '"' in text:
        return double_ord
    
    if fancy.rq in text:
        return fancy
    
    return single
```
