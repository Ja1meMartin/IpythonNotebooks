# What Relationships Can We Store
- Refernce : Data is stored in 2 different collections such as User and (blog) Post. The latter, the many, references the former.

```python
class Course(Document):
    nsme = StringField()
    

    
    
class Horse(Document):
    name = StringField(required=True)
    
class Race(Document):
    name = StringField(required=True)
    meta = {'alloe_inheritance':True}
    
class Meeting(Document):
    course = ReferenceField(Course)
    date = DateTimeField(required=True)
    races = ListField(ReferenceField(Race))
    
    


    
```
