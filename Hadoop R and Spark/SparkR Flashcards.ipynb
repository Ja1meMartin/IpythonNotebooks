```python
from bs4 import BeautifulSoup

import requests

base = "https://spark.apache.org/docs/1.5.0/api/R/"
url = "00Index.html"

r  = requests.get(base+url)

data = r.text

soup = BeautifulSoup(data)
```

```python
links = []
for link in soup.find_all('a')[27:]:
    if link.get('href'): 
        print "---",link.get('href'),"---"
        links.append(link.get('href'))
```

```python
new_url = links[5]

def get_text(new_url):
    r  = requests.get(base+new_url)

    data = r.text

    soup = BeautifulSoup(data)
    
    return '"{}"!!"{}";'.format(new_url[:-5],soup.text)

csv = "\n".join(get_text(new_url) for new_url in links)
```

```python
with open("sparkR.csv", 'w') as csv_file:
    csv.replace("\n","")
    csv_file.write(csv)
```
