```python
address = "http://www.espnfc.com/barclays-premier-league/23/table"

import urllib2
from bs4 import BeautifulSoup

url = urllib2.urlopen(address)
content = url.read()
soup = BeautifulSoup(content)
```

```python
team_links = [c.find('a', href = True) for c in soup.find_all('td') if c.has_attr('class') and 'team' in c.get('class') ]

#print len(team_links), team_links

for t in team_links:
    print t.text, t['href'][:-5]
    
```

```python
team = team_links[0]['href'][:-5] + 'squad'
url = urllib2.urlopen(team)
content = url.read()
soup = BeautifulSoup(content)
```

```python
tables = soup.findAll('table')
print len(tables)

gk_table = tables[0]

goalkeepers = [gk for gk in gk_table.findAll('tr') if  not gk.has_attr('class')]
for g in goalkeepers:
    
    gk_tag = g.find("td",'pla')
    gk = str(gk_tag)
    gk_split = gk.split("player/")
                        
    end = gk_split[1].index("/")
    gk_id = int(gk_split[1][:end].strip())
    print "ID",gk_id, "---"
    gk_name =  gk_tag.text.strip()
    print "NAME",gk_name
    print "\n------------\n"
    
"""
    Find id
    player name
    club
"""
```
