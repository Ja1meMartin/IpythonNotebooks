```python
%pylab inline
from matplotlib.pyplot import *
import numpy as np
from sympy import *
from sympy.interactive import printing
import functools
from sympy.solvers import solve
from __future__ import division
```

```python
x = Symbol('x')

def polyFromCoeff(coeff):
    
    def sign(pow, co):
        if co == 0:
            return ""
        elif co < 0:
            return "%s*x**%s" %( str(co), str(pow) )
        else:
            return "+%s*x**%s" %( str(co), str(pow) )
        
    p = " ". join ([sign(a,b) for a,b in coeff])
        
    
        
    return Poly(p)

def coeffFromPoly(poly):
    return [(a[0],b) for a,b in poly.as_dict().items()]
```

```python

    

class Function(object):
    def __init__(self, coeff, start=-6, end = 6, points = 20, function = None):
        # more points = more accurate but take more time
        self.points =int((end-start)* points)
        self.start = start
        self.end = end
        
        
        # tuples in the form (coefficient, power of x)
        length = len(coeff)
        
        
        # function
        if function:
            self.function = function
            
        else:
            coeff = [(length - 1 - index, value) for index, value in enumerate(coeff)]
            
            self.function = polyFromCoeff(coeff)
            
            
              
            
        #self.coeff = [(length - 1 - index, value) for index, value in enumerate(coeff)]
        self.coeff = coeffFromPoly(self.function)
      
        
        self.X = np.linspace(start, end, self.points)
        
        
        # coefficients for the first and second derivatives
        #self.D1 = [(a*b, b-1) for a,b in self.coeff[:-1]]
        #self.D2 = [(a*b, b-1) for a,b in self.D1[:-1]]
        
        self.D1 = self.function.diff(x)
        self.D2 = self.D1.diff(x)
        
        # set the Y values
        self.Y = np.zeros(self.points)
        for b,a in self.coeff:  
            
            self.Y = self.Y +a*(self.X**b) 
            
            
       
    def y(self,coeff, change = None):
        y = np.zeros(self.points)
        
        
        for b,a in coeff:
            if b == change:
                
                y = y -a*(self.X**b)
            else:
                
                y = y +a*(self.X**b)
            
        return y
        
        
    def values(self):
        pass
    
    def d1(self):
        return self.D1
    
    def d2(self):
        return self.D2
    
    
    # plot the function and its first derivative
    def plotD1(self):
        def second_derivative_test(tup):
            mid = tup
            if self.D2(mid) > 0:
                print mid, "green", self.D2(mid)
                return -1
            elif self.D2((mid)) == 0:
                return 0
            else:
                print mid, "red", self.D2(mid)
                return 1
            
            
        
        colours = {1:'g', -1:'r', 0:'y'}
        
        solutions = sorted( solve(self.D1,x))
        
        print "solutions", solutions
        
        
        tupes = []          
        
        
        if len(solutions) ==  1:
            second_derivative = second_derivative_test(solutions[0])
            #tupes = [(self.start, solutions[0]), (solutions[-1], self.end)]
            plt.axvspan(self.start, solutions[0], color=colours[second_derivative], alpha=0.2)
            plt.axvspan(solutions[-1], self.end, color=colours[second_derivative], alpha=0.2)
        
        if len(solutions) > 1:
            second_derivative_start = -second_derivative_test(solutions[0])
            second_derivative_end = second_derivative_test(solutions[-1])
            #tupes = [(self.start, solutions[0]), (solutions[-1], self.end)]
            plt.axvspan(self.start, solutions[0], color=colours[second_derivative_start], alpha=0.2)
            plt.axvspan(solutions[-1], self.end, color=colours[second_derivative_end], alpha=0.2)
            
            tupes.extend([(sol, solutions[i+1], second_derivative_test(sol)) for i, sol in list(enumerate (solutions[:-1]))])
            
            for t in tupes:
                col = colours[second_derivative_test(t[0])]
                print t,"col", col
                plt.axvspan(t[0], t[1], color=col, alpha=0.2)
            
            
            
        plt.axhline(color='k')
        
        d1coeff = coeffFromPoly(self.D1)
        
        
        pyplot.plot(self.X,self.Y)
        
         
        pyplot.plot(self.X,self.y(d1coeff))
    
    # plot the function and its first AND second derivatives
    def plotD1D2(self):
        d1coeff = coeffFromPoly(self.D1)
        d2coeff = coeffFromPoly(self.D2)
        plt.axhline(color='k')
        pyplot.plot(self.X,self.Y)
        pyplot.plot(self.X,self.y(d1coeff))
        pyplot.plot(self.X,self.y(d2coeff))
        grid(True)
    
    def plot(self, coeff = None):
        pyplot.plot(self.X,self.Y)
        plt.axhline(color='k')
        pyplot.grid()
            
        pyplot.show()
        
    def change(self,co):
        plt.axhline(color='k')
        pyplot.plot(self.X,self.Y)
        #pyplot.plot(self.X,self.y(self.D1))
        #pyplot.plot(self.X,self.y(self.D1,co-1))
        pyplot.grid()
        pyplot.plot(self.X,self.y(self.coeff, co))
        
    def changeWithDerivative(self,co):
        x1 = self.X
        d1coeff = coeffFromPoly(self.D1)
        plt.axhline(color='k')
        #plt.subplot(2, 1, 1)
        plt.plot(x1, self.Y, 'b-', label="original")
        plt.plot(x1, self.y(d1coeff), 'b-')
        
        plt.grid(which='major', color='k', linestyle='-')

        #plt.subplot(2, 1, 2)
        plt.plot(self.X,self.y(self.coeff, co), 'r-',label="changed")
        plt.plot(self.X, self.y(d1coeff,co-1), 'r-')
        
        
        plt.grid(which='major', color='k', linestyle='-')
        plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

        plt.show()
    
             

def functionFromRoots(listOfRoots, start=-6, end = 6, points = 10):
    factors = [(x-a) for a in listOfRoots]
    print "Here are the factors:", factors
       
    def ffr(a,b):
        return expand(a*b)
    poly = functools.reduce(lambda a,b: ffr(a,b), factors)
    
    print "Here is the polynomial", latex(poly)
    
    poly = Poly(poly)
    
    
    
    
    return Function([4], start, end, points, function = poly)
    
  
    



#a.plotD1D2()
```

### Change The Sign Of a Coefficient
What happens if we change the sign for each part. First lets plot a function.  
  
Note that you can change the numbers here and then go 'Cell' > 'Run All Below' to change everything in one go.

```python
m = Function([2,-1,2])
m.plot()
```

Ok let's change the sign by calling the power of the exponent of the cell we wish to change.  
m.change(0) changes the constant, m.change(1) changes the coeff of the x ... where m is the name of our function

```python
m.change(0)
```

All the constant does is move a curve up or down. Note it may look like the curves are getting closer together but if you check where they cross grid lines you will see that they are not.

Next up change the x part.

```python
m.change(1)
```

Changing the x part will move the graph either left or right. Can you figure out which?  

Hint: If the coefficient of x is positive then it is pulling the graph down until it reaches 0. After 0 it is pulling the graph up. The converse is true for a negative coefficient.

Perhaps it is easier to see with derivatives.

```python
m.changeWithDerivative(1)
```

Changing the sign of the x part will change the constant part of the derivative. We have already learned that this shifts things up and down. 

So if changing the constant moves us vertically and changing the x part moves us horizontally then what is left for the x^2 part?

```python
m.changeWithDerivative(2)
```
