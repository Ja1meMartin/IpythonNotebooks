```python
#import lxml.etree
#import lxml.html
import urllib2
from bs4 import BeautifulSoup

match_number = 395691

def get_comments(match_number):
    address = "http://www.espnfc.com/gamecast/" + str(match_number) + "/gamecast.html"

    url = urllib2.urlopen(address)
    content = url.read()
    return BeautifulSoup(content)
    
soup = get_comments(match_number)    
```

```python
comments = [c for c in soup.find_all('li') if c.has_attr('id') and c.get('id').startswith(str(match_number))]

```

```python
text_comments = [(c.text, c.get('id')) for c in comments]
incidents = []

def incident(comment_id_tuple, inList, outList):
    #print "comment", comment
    
    comment, c_id = comment_id_tuple
    
    if comment.startswith("Foul"):
        outList.append( ("Foul", comment, c_id))
        #print "Foul", comment
        comment, c_id = inList.pop()
        lines = comment.split('\n')
        #print "next", "Fouled", lines[2]
        outList.append(("Fouled", lines[2], c_id))
        return
    
    if comment.startswith("Hand "):
        outList.append( ("Hand Ball", comment, c_id))
        return
    
    if comment.startswith("Corner"):
        outList.append( ("Corner", comment, c_id))
        return
    
    if comment.startswith("Attempt"):
        outList.append( ("Attempt", comment, c_id))
        return
    
    if comment.startswith("Goal"):
        outList.append( ("Goal", comment, c_id))
        return
    
    if comment.startswith("Offside"):
        outList.append( ("Offside", comment, c_id))
        return
    
    if comment.startswith("Substitution"):
        outList.append( ("Sub", comment, c_id))
        return
    
    if comment.startswith("Delay"):
        outList.append( ("Delay", comment, c_id))
        return
    
    if comment.startswith("Penalty"):
        outList.append( ("Penalty", comment, c_id))
        return
    
    if comment.startswith("First") or comment.startswith("Second"):
        return
    
    if comment.startswith("Own G"):
        outList.append( ("OG", comment))
        return
    
    if "wins a fr" in comment:
        outList.append( ("Fouled", comment, c_id))
        #print "Foul", comment
        comment, c_id = inList.pop()
        lines = comment.split('\n')
        #print "next", "Fouled", lines[2]
        outList.append(("Foul", lines[2], c_id))
        return
    
    if "yellow card" in comment:
        outList.append(("Yellow", comment, c_id))
        return
        
    if "red card" in comment:
        outList.append(("Red", comment, c_id))
        return
        
    if "hits the" in comment:
        outList.append(("Woodwork", comment, c_id))
        return
    
    
    
        
        
    print comment[0:9], "comment ??", comment
    outList.append (("NO INCIDENT FOUND", comment))
    
while text_comments:
    comment = text_comments.pop()
    
    lines = comment[0].split('\n')
    if len(lines[1]) > 1:
        time =  int(lines[1][:-1])
        #print time
        #print "-->", lines[2]
        incident((lines[2], comment[1]), text_comments, incidents)
        #print "\n----------\n"


    
```
